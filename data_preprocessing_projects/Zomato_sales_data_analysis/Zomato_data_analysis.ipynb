```python
# This command is used to check for missing (null) values in a Pandas DataFrame.
# Breaking it Down:
# 1. df.isnull()
#     -> This returns a DataFrame of the same shape as df, where each value is:
#             ->> True if the value is null (NaN or None).
#             ->> False if the value is not null.
# 2. .sum()
#     ->Since True is treated as 1 and False as 0, summing along columns gives the total number of missing values per column.
df.isnull().sum()
```

```python
# this for loop will iterate through the column and find the number of columns which have 
# total number of null values greater than 0
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

## Numverical variables

```python
final_df.columns
```

```python
final_df.groupby(["Aggregate rating",'Rating color', 'Rating text']).size().reset_index()
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
rating
```

## Observation
1. when Rating is between 4.5 to 4.9 --->Excellent
2. when Rating is between 4.0 to 4.4 --->Very Good
3. when Rating is between 3.5 to 3.9 --->Good
4. when Rating is between 2.5 to 3.4 --->Average
5. when Rating is between 0.0 to 2.4 --->poor

```python
rating.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Aggregate rating", y="Rating Count",data=rating)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Aggregate rating", y="Rating Count",hue="Rating color",data=rating,palette=
           ["blue","red","orange","yellow","green","green"])
```

## obserations:
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x="Rating color",data=rating,palette=
           ["blue","red","orange","yellow","green","green"])
```
